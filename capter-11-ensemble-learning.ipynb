# Ensemble Learning

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
```

```python
x,y=make_classification(n_samples=2000,n_features=10,n_informative=8,n_redundant=2,random_state=11)
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=11)
```

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
cross_val_score(gnb,x_train,y_train,cv=3).mean()
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
cross_val_score(lr,x_train,y_train,cv=3).mean()
```

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
cross_val_score(rf,x_train,y_train,cv=3).mean()
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
cross_val_score(dt,x_train,y_train,cv=3).mean()
```

```python
#voting
vc=VotingClassifier([('NavieBayes',gnb),('LogisticRegression',lr),('DecisionTreeClassifier',dt)])
```

```python
cross_val_score(vc,x_train,y_train).mean()
```
