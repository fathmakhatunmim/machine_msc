# Bias-Variance Tradeoff

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('/kaggle/input/malborn/Melbourne_housing_FULL (1).csv')
```

```python
data.head(2)
```

```python
data.nunique()
```

```python
data.isnull().sum()
```

```python
col=['Propertycount','Distance','Bedroom2','Bathroom','Car']
data[col]=data[col].fillna(0)
```

```python
data.isnull().sum()
```

```python
data['Landsize']=data['Landsize'].fillna(data['Landsize'].mean())
```

```python
data.isnull().sum()
```

```python
data['BuildingArea']=data['BuildingArea'].fillna(data['BuildingArea'].mean())
```

```python
data.isnull().sum()
```

```python
# data.dropna(subset=['YearBuilt', 'CouncilArea', 'Lattitude', 'Longtitude', 'Regionname'], inplace=True)
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data.head(2)
```

```python
data.sample(5)
```

```python
data=pd.get_dummies(data,drop_first=True)
```

```python
data.shape
```

```python
x=data.drop('Price',axis=1)
```

```python
y=data['Price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
data.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)*100
```

```python
reg.score(x_test,y_test)*100
```

```python
reg. coef_
```

```python
from sklearn.linear_model import Ridge
rd=Ridge(alpha=20)
rd.fit(x_train,y_train)
```

```python
rd.score(x_test,y_test)
```

```python
rd.score(x_train,y_train)
```

```python
from sklearn.linear_model import Lasso
la=Lasso()
la.fit(x_train,y_train)
```

```python
la.score(x_train,y_train)
```

```python
la.score(x_test,y_test)
```

```python
from sklearn.linear_model import ElasticNet
el=ElasticNet()
el.fit(x_train,y_train)
```

```python
el.score(x_train,y_train)
```

```python
el.score(x_test,y_test)
```
