```python
import pandas as pd 
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
df = pd.read_csv('/kaggle/input/machine-4th/homeprices.csv')
df
```

```python
df.head(2)
```

```python
df.tail(1)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.bedrooms.median()
df.bedrooms=df.bedrooms.fillna(df.bedrooms.median())
df.bedrooms
```

```python
inputs=df.drop('price',axis='columns''')
inputs
```

```python
price=df.price
price
```

```python
reg=linear_model.LinearRegression()
reg.fit(inputs,prices)
```

```python
reg. coef_
```

```python
reg. intercept_
```

```python
reg.predict([[3000,3,40]])
```

```python
y_pred=reg.predict(inputs)
print('MAE=',mean_absolute_error(price,y_pred))
print('MsE=',mean_squared_error(price,y_pred))
print('MsE=',np.sqrt(mean_squared_error(price,y_pred)))

print('R2_score=',r2_score(price,y_pred))

```

# Home Work

```python
data= pd.read_csv('/kaggle/input/homework/capter_4(hw).csv')
data
```

```python
print(data.columns.tolist())
```

```python
x = data[['experience', ' testscore', 'interviewscore']]
y = data['salary']
```

```python
from sklearn import linear_model
model =linear_model.LinearRegression()
model.fit(x,y)
```

```python
candidates =pd.DataFrame({
    'experience':[2,12],
    ' testscore':[9,10],
      'interviewscore':[6,10]
})
```

```python
prediction=model.predict(candidates)
# prediction
```

```python
for i, salary in enumerate(prediction):
    print(f"Candidate {i+1} expected salary: {salary:.2f}")
```
