# Confusion Matrix for Dog Image Recognition with Numbers
 Implementaion with python

```python
import pandas as pd
import numpy as np
from sklearn. metrics import confusion_matrix,classification_report
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
actual=np.array(['dog', 'dog', 'dog' ,'Not_dog', 'dog', 'Not_dog', 'dog' ,'dog','Not_dog','Not_dog',])
predict=np.array(['dog','Not_dog','dog','Not_dog','dog','dog','dog','dog','Not_dog','Not_dog',])
```

```python
actual
```

```python
predict
```

```python
cm= confusion_matrix(actual,predict)
cm
```

```python
print(classification_report(predict,actual))
```

# Confusion Matrix for Dog Image Recognition with Numbers
 Implementaion with python

```python
sns.heatmap(cm,annot=True,fmt='.0f',xticklabels=['1','0'],yticklabels=['dog','Not_dog'])
plt.ylabel('Actual',fontsize=12)
plt.title('confusion_matrics',fontsize=17,pad=22)
plt.gca().xaxis.set_label_position('top')
plt.xlabel('Prediction', fontsize=13)
plt.gca().figure.subplots_adjust(bottom=0.2)
plt.gca().figure.text(0.4,0.05,'prediction',ha='center',fontsize=13)
plt.show()
```

# Prediction of Insurance buying based on age

```python
data = pd.read_csv('/kaggle/input/insura/insurance_data.csv')
data
```

```python
plt.scatter(data.age,data.bought_insurance)
```

#  #Separate Feature and Traget matrixs

```python
x= data.iloc[:,:-1].values
y=data.iloc[:,1].values
```

```python
x
```

```python
y
```

```python

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
y_train
```

```python
ml=linear_model.LogisticRegression()
ml.fit(x_train,y_train)
```

```python
predict=ml.predict(x_train)
predict
```

```python
plt.scatter(x_train,y_train,marker='*',color='red')
plt.scatter(x_train,predict,marker='.',color='blue')
plt.xlabel("Age of person")
plt.ylabel("Bought Insurance 1=Bought 0=Did not Buy")
```
