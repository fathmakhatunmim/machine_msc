#  Column Transformer

```python
from sklearn.compose import ColumnTransformer
```

```python
trans=ColumnTransformer(
    transformers=[ 
    ('tr1',SimpleImputer(),['fever']),
    ('tr2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tr3',OneHotEncoder(drop='first',sparse_output=False),['gender','city'])
    ],remainder='passthrough')
```

```python
trans
```

```python
trans.fit(x_train)
```

```python
x_train_col_transformed=trans.transform(x_train)
```

```python
x_train_col_transformed.shape
```

```python
x_train_transformed=pd.DataFrame(x_train_col_transformed,columns=['fever','cough','gender','Kolkata','Bangalore','Delhi','age'])
```
