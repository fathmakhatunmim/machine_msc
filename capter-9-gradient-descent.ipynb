#  Gradient Descent

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
x,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(4,3))
plt.scatter(x,y)
```

```python
reg=LinearRegression()
reg.fit(x,y)
```

```python
m=reg.coef_
```

```python
m
```

```python
b=reg.intercept_
```

```python
b
```

```python
y_pred=reg.predict(x)
```

```python
plt.figure(figsize=(4,3))
plt.scatter(x,y)
plt.plot(x,y_pred,color='red')
```

```python
b=-100
m=78.35
lr=0.01
n=1

for i in range(n):
    slope=-2*np.sum(y-m*x.ravel()-b)
    b=b-(lr*slope)
    y_pred_new=m*x+b
    print(b,slope)
    plt.figure(figsize=(4,3))
    plt.plot(x,y_pred_new)
    
plt.plot(x,y_pred)    
plt.scatter(x,y)
```

```python
b=-100
m=78.35
lr=0.01
epoch=4
plt.figure(figsize=(4,3))
for i in range(epoch):
    slope=-2*np.sum(y-m*x.ravel()-b)
    b=b-(lr*slope)
    y_pred_new=m*x+b
    print(b,slope)
    # plt.figure(figsize=(4,3))
    plt.plot(x,y_pred_new)
    
plt.scatter(x,y)    
plt.plot(x,y_pred,color='red')    
plt.scatter(x,y)
```
