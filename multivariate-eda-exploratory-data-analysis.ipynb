# numerical to numerical

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# tips=sns.load_dataset('tips')
tips=pd.read_csv('/kaggle/input/tips-dataset/tips.csv')
```

```python
tips
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='smoker')
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='smoker',size='size')
```

# numerical to categorical

```python
data=pd.read_csv('/kaggle/input/titanicda/titanic.csv')
```

```python
data.head(2)
```

```python
print(data.columns)
```

```python
#age-numerical pclass-catagorical
plt.figure(figsize=(4,4))
sns.barplot(data=data,x='Pclass',y='Age')
```

```python
plt.figure(figsize=(4,4))
sns.barplot(data=data,x='Pclass',y='Age',hue='Sex')
```

```python
plt.figure(figsize=(4,4))
sns.boxplot(data=data,x='Sex',y='Age')
```

```python
plt.figure(figsize=(4,4))
sns.boxplot(data=data,x='Sex',y='Age',hue='Survived')
```

```python
x=data[data['Survived']==0]
```

```python
y=data[data['Survived']==1]
```

```python
x.head(2)
```

```python
y.head(2)
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=x,x='Age')
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=x,x='Age')
sns.kdeplot(data=y,x='Age')
```

```python
x=data[(data['Survived']==0)&(data['Sex']=='male')]
```

```python
x.head(3)
```

```python
y=data[(data['Survived']==0)&(data['Sex']=='female')]
```

```python
y
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=x,x='Age')
sns.kdeplot(data=y,x='Age')
```

# categorical to categorical

```python
cm=pd.crosstab(data['Pclass'],data['Survived'])
```

```python
sns.heatmap(cm,annot=True,fmt='.0f')
```

```python
cm
```

```python
cm=pd.crosstab(data['Parch'],data['Survived'])
```

```python
cm
```

```python
flight=pd.read_csv('/kaggle/input/flights/flights.csv')
```

```python
flight
```

```python
flight.groupby('month').sum()['passengers']
```

```python
fx=flight.pivot(index='year',columns='month',values='passengers')
fx
```

```python
sns.heatmap(fx,annot=True,fmt='.0f')
```

```python
px=flight.query("month =='May'")
sns.lineplot(data=px,x='year',y='passengers')
```
